### COVID Datases

```python
# Here I am importing a World Dataset from https://data.humdata.org/dataset/novel-coronavirus-2019-ncov-cases?force_layout=desktop
# I saved this dataset in the local folder

df = pd.read_csv('time_series_covid19_confirmed_global.csv')
```

```python
# As we can see, some countries have a discrete row and other (feel free to see this for yourself) are broken down into provinces.
df.head()
```

```python
# I will add all desceate countries in another dataset and then I will "compress" the countries with multiple regions into one country
# Here we can see all the unique countries.
df.iloc[:,1].unique()

# Now I added the multiple rows with the same country into one row.
df = df.groupby("Country/Region").sum()
```

```python
# Here I showed the new dataset (feel free to take away ".head()" to see the full dataset)
# Latitude and Longitude get messed up doing the past sum, but I won't use them eitherway.
df.head()
```
