```python
# We will be using Pandas for reading al the relevant datasets.
import pandas as pd
```

```python
# Here we read the data from the Energy Dataset
df = pd.read_excel('CrudeOil.xls',header=None)
# We make a time range
time = list(range(1973, 2019))
```

```python
# Now we turn the data range to a pandas readable format
TimeList = pd.DataFrame(time)
```

```python
# Now we make a list with all the countries from the original datast
countries = []
for i in range(0,len(df[0])):
    if i%8 == 0:
        countries.append(df[0][i])
```

```python
# Here we can see the origianl dataset. We can see that its very messy
df
```

```python
# Now we filter and only retrived the data for oil production for each country that we have in our "Country" list.
ProductionByCountry = []
for i in range(0,len(df[0])):
    if df[0][i] == "        Total petroleum and other liquids (Mb/d)":
        ProductionByCountry.append(df.iloc(0)[i])
```

```python
# Now we concatonate the lists

df_test = ProductionByCountry[0]
for i in range(1,len(ProductionByCountry)):
    df_test = pd.concat([df_test,ProductionByCountry[i]],axis=1) 
```

```python
# Here we not transpose the data and drop a few unnecesary columns
Meh = df_test.T
Meh = Meh.drop([1,2,3,4,5,6,7], axis=1)
```

```python
#Now we include our country list into our other dataset
CountriesList = pd.DataFrame(countries)
Meh['Countries'] = CountriesList
```

```python
# Here is our new dataset
Meh
```

```python
# Now we export all three datasets and then in Excel/Flourish add them together to make the visuals.
Meh.to_csv('test.csv',index=False)
```
