## Python code to create Top 10 Most Successful Movies (1918-2020)


```python
# We will be using Pandas for reading al the relevant datasets.
import pandas as pd
```

```python
# Here we read the data from the Energy Dataset
df = pd.read_excel('Uncleaned Data.xlsx')
# We make a time range
```

```python
df = df[df.Title.notnull()]
```

```python
df = df[df["Worldwide gross"].notnull()]
```

```python
df = df[df["Year"].notnull()]
```

```python
df.head()
```

```python
int(min(df["Year"]))
```

```python
timelist = list(range(int(min(df["Year"])),int(max(df["Year"]))+1))
```

```python
MovieYearRelease = []

FullGrossList = []


for i in range(len(timelist)):
    yearList = []
    for j in range(len(df)):
        if int(df.iloc[j,0]) <= timelist[i]:
            yearList.append(df.iloc[j,2])
            if int(df.iloc[j,0]) == timelist[i]:
                MovieYearRelease.append(df.iloc[j,1])
        else:
            yearList.append(0)
    FullGrossList.append(yearList)
                
                
            
```

```python
MovieYearRelease
```

```python
FullGrossList
```

```python
dfObj = pd.DataFrame(FullGrossList).T 
```

```python
dfObj["Movies"] = MovieYearRelease
```

```python
dfObj
```

```python
# Now we export all three datasets and then in Excel/Flourish add them together to make the visuals.
dfObj.to_csv('MoviesList.csv',index=False)
```
